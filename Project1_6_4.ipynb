## Defining Function for calculating MAPE

```python
def mape (actual, predicted):
  actual, predicted = np.array (actual), np.array (predicted)
  return np.mean (np.abs ( (actual - predicted) / actual)) * 100
```

### Defining X and y data and doing train test split

```python
x_data = df_final.drop(columns =['co2_emi'])
```

```python
y_data= df_final['co2_emi']
y_data
```

```python
# Step 1 -> train/test/split
X_train, X_test,y_train,y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
X_train
```

```python
y_train
```

## Calculating VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

#add_constant in above code comes from the the below link
#https://stackoverflow.com/questions/42658379/variance-inflation-factor-in-python

#x_required = X_train.assign(const=1)
x_required = X_train.assign(const=1)
x_required

vif = []
for i in range (x_required.shape[1]):
    vif.append(variance_inflation_factor(x_required, i))

pd.DataFrame({'vif': vif}, index=x_required.columns[0:150])
```

## Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

model2 = DecisionTreeRegressor()
model2.fit(X_train, y_train)

DTscore = model2.score(X_test,y_test)

predictions2 = model2.predict(X_test)

mae2 = mean_absolute_error(y_test, predictions2)
rmse2 = np.sqrt(mean_squared_error(y_test, predictions2))
print ('Decision Tree Regressor::')
print(f'MAE    = {mae2:.2f}')
print(f'RMSE   = {rmse2:.2f}')
print('MAPE   =', 100-mape(y_test,predictions2))
print(f'R^2    = {DTscore:.2f}')
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(X_train, y_train)

predictions_rf = model_rf.predict(X_test)

mae_rf = mean_absolute_error(y_test, predictions_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, predictions_rf))

r2_rf = r2_score(y_test, predictions_rf)
print ('Random Forest Regressor::')
print(f'MAE    = {mae_rf:.2f}')
print(f'RMSE   = {rmse_rf:.2f}')
print ('MAPE   =',100-mape(y_test,predictions_rf))
print(f'R^2    = {r2_rf:.2f}')
```

## Using Neural Networks

```python
import tensorflow as tf

model3 = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])

model3.compile(optimizer='adam', loss='mean_squared_error')

model3.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

loss = model3.evaluate(X_test, y_test)
print(f'Test loss: {loss}')

predictions3 = model3.predict(X_test)
r2_3 = r2_score(y_test, predictions3)

mae3 = mean_absolute_error(y_test, predictions3)
rmse3 = np.sqrt(mean_squared_error(y_test, predictions3))

print ('Neural Networks::')
print(f'MAE    = {mae3:.2f}')
print(f'RMSE   = {rmse3:.2f}')
print('MAPE   =', 100-mape(y_test,predictions3))
print(f'R^2    = {r2_3*100:.2f}')
```

## Linear Regression

```python
model = sm.OLS(y_train,X_train).fit()
```

```python
y_pred= model.predict(X_test)

mae= mean_absolute_error(y_test, y_pred)
mse= mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print ('Linear Regression::')
print('MAE     = ',mae)
print('RMSE    = ',rmse)
print('MAPE    = ', 100-mape(y_test,y_pred))
print('R^2     = ', model.rsquared)
print('Adj R^2 = ', model.rsquared_adj)
```
